### tkinter

```python
# close window after 1 second
from tkinter import *

tk = Tk()
tk.geometry("200x100+0+0") # (width x height + x + y)
tk.title("title")

tk.after(1000, tk.destroy) # close window after 1 second
tk.lift()
tk.focus_force()
tk.mainloop()
```

```python
# trigger callback per 300 ms
from tkinter import *

def f1(*args):
    tk = args[0]
    print("f1")
    tk.after(300, f1, tk) # trigger callback per 300 ms

tk = Tk()
tk.geometry("100x100+0+0")
tk.after(300, f1, tk) # trigger callback per 300 ms

tk.after(1000, tk.destroy)
tk.lift()
tk.focus_force()
tk.mainloop()
```

```python
from tkinter import *
import math

tk = Tk()
c = Canvas(tk, width=400, height=400)
c.pack()
x_center, y_center, r = 200, 200, 150
x, y = [], []
for i in range(12):
    x.append(x_center + r * math.cos(i*math.pi/6))
    y.append(y_center + r * math.sin(i*math.pi/6))
for i in range(12):
    for j in range(12):
        c.create_line(x[i],y[i],x[j],y[j])
        
tk.after(1000, tk.destroy)
tk.lift()
tk.focus_force()
mainloop()
```

```python
from tkinter import *
def f1(source):
    print("f1", source)
    color = "#ffff%02x" % int(source)
    c.config(bg=color)
    
tk = Tk()
c = Canvas(tk, width=640, height=240)
s = Scale(tk, from_=0, to=255, command=f1)
s.set(150)
s.grid(row=1, column=1)
c.grid(row=2, column=1, columnspan=2)

tk.after(10000, tk.destroy)
tk.lift()
tk.focus_force()
mainloop()
```

```python
from tkinter import *

def keyboard_event(event):
    # cannot listen multiple keys
    if event.keysym == "Left":
        c.move(o1, -5, 0)
    elif event.keysym == "Right":
        c.move(o1, 5, 0)
    elif event.keysym == "Up":
        c.move(o1, 0, -5)
    elif event.keysym == "Down":
        c.move(o1, 0, 5)
    elif event.keysym == "Escape":
        tk.destroy()
    else:
        print(event)

tk = Tk()
c = Canvas(tk, width=300, height=300)
c.pack()
c.bind_all('<KeyPress>', keyboard_event) # listen all keys

o1 = c.create_oval(100, 100, 150, 150, fill='yellow', outline='lightgray')

tk.after(10000, tk.destroy)
tk.lift()
tk.focus_force()
mainloop()
```

### Keyboard, Mouse Event

```python
from tkinter import *
from PIL import ImageTk, Image

window_width = 600
window_height = 480
x, y = None, None

def keyEvent(event):
    if event.keysym == "Escape":
        tk.destroy()

def mouseMoveEvent(event):
    global x, y
    canvas.create_rectangle(event.x-2, event.y-2, event.x + 2, event.y + 2, width=1, fill="green")
    if x is not None:
        canvas.create_line(x, y, event.x, event.y, fill="yellow", width=1)
    x = event.x
    y = event.y

def mouseClickEvent(event):
    canvas.create_rectangle(event.x-5, event.y-5, event.x + 5, event.y + 5, width=1, fill="red")

def clearImg():
    canvas.delete("all")

tk = Tk()
canvas = Canvas(tk, width=window_width, height=window_height)
canvas.pack()

tk.title("Game ver=0.01")
tk.geometry("%dx%d+0+0" % (window_width, window_height))
tk.bind_all("<KeyPress>", keyEvent)
tk.bind_all("<KeyRelease>", keyEvent)
tk.bind_all("<Motion>", mouseMoveEvent)
tk.bind_all("<Button>", mouseClickEvent)

img = Image.open("res/mario.png")
img = img.resize((200, 150), Image.ANTIALIAS)
img = ImageTk.PhotoImage(img)
canvas.create_image(10, 10, image=img, anchor=NW)

tk.lift()
tk.focus_force()
#tk.after(2000, clearImg)
#tk.after(10000, tk.destroy)
tk.mainloop()
```
