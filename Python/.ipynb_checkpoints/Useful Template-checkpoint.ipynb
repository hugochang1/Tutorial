### logging

```python
import logging
from logging import debug as log
from logging import warning as logw
from logging import error as loge

def logging_init():
    #reset logging configuration
    for handler in logging.root.handlers[:]:
        logging.root.removeHandler(handler)

    logging.basicConfig(level=logging.DEBUG, #filename="log.txt",
                        format="%(asctime)s.%(msecs)03d %(name)-12s %(levelname)-8s: %(message)s",
                        datefmt='%Y-%m-%d %H:%M:%S')
```

```python
logging_init()

log("log")
logw("logw")
loge("loge")
```

### plot

```python
import matplotlib.pyplot as plt
from datetime import datetime

# datas
date = ['2017/1/1', '2017/1/2', '2017/1/3', '2017/1/4', '2017/1/5', '2017/1/6', '2017/1/7', '2017/1/8', '2017/1/9', '2017/1/10', '2017/1/11', '2017/1/12', '2017/1/13', '2017/1/14', '2017/1/15', '2017/1/16', '2017/1/17', '2017/1/18', '2017/1/19', '2017/1/20', '2017/1/21', '2017/1/22', '2017/1/23', '2017/1/24', '2017/1/25', '2017/1/26', '2017/1/27', '2017/1/28', '2017/1/29', '2017/1/30', '2017/1/31']
a1 = [26, 25, 22, 27, 25, 25, 26, 22, 18, 20, 21, 22, 18, 15, 15, 16, 23, 23, 22, 18, 15, 17, 16, 17, 18, 19, 24, 26, 25, 27, 18]
a2 = [20, 18, 19, 20, 19, 20, 20, 18, 17, 16, 18, 18, 14, 12, 13, 13, 16, 18, 18, 12, 12, 12, 13, 14, 13, 13, 13, 16, 17, 14, 14]

# covert date from string to datetime
dates = []
for d in date:
    dates.append(datetime.strptime(d, "%Y/%m/%d"))

# figure size
fig = plt.figure(dpi=80, figsize=(12,4))

# plot
a, = plt.plot(dates, a1, "-*", linewidth=2, label="label A")
b, = plt.plot(dates, a2, "-o", label="label B")
plt.fill_between(dates, a1, a2, color="y", alpha=0.1) # fill in the color between two lines

plt.legend(handles=[a, b], loc="upper left", bbox_to_anchor=(1,1))

# x, y ticks
plt.axis([dates[0], dates[30], 0, 40]) # set x and y range (x1, x2, y1, y2)
plt.xticks(dates, date)
plt.yticks([0,5,10,15,20,25,30])

plt.title("Title", fontsize=24)
plt.xlabel("Xlabel", fontsize=14)
plt.ylabel("Ylabel", fontsize=14)
plt.tick_params(axis="both", labelsize=10, color="red")

plt.grid(True)
plt.tight_layout(pad=1)
fig.autofmt_xdate(rotation=30)
plt.savefig("test/image0.jpg", bbox_inches="tight")
plt.show()
```

### email

```python
import smtplib
from email.mime.application import MIMEApplication
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

from_addr = "sand208109@gmail.com"
pwd = "helloworld123"
to_addr_list = ["sand208109@gmail.com"]

msg = MIMEMultipart()
msg["From"] = "sand208109@gmail.com"
msg["To"] = "sand208109@gmail.com"
msg["Cc"] = "sand208109@gmail.com"
msg["Subject"] = "subject misc"

msg.attach(MIMEText("this is my content"))

with open("res/image.jpg", "rb") as f:
    data = f.read()
    part = MIMEApplication(data)
    part['Content-Disposition'] = 'attachment; filename="phone.jpg"'
    msg.attach(part)

with open("res/music.mp3", "rb") as f:
    data = f.read()
    part = MIMEApplication(data)
    part['Content-Disposition'] = 'attachment; filename="music.mp3"'
    msg.attach(part)

try:
    smtp = smtplib.SMTP("smtp.gmail.com", 587)
    smtp.ehlo()
    smtp.starttls()
    smtp.login(from_addr, pwd)
    status = smtp.sendmail(from_addr, to_addr_list, msg.as_string())
    if status == {}:
        print("mail sent successfully")
    smtp.quit()
except Exception as err:
    print(err)
```
