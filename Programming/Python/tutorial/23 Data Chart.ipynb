### simple line

```python
import matplotlib.pyplot as plt

data = [1,4,9,16,25,36,49,64]
plt.plot(data, linewidth=3)
plt.title("my title", fontsize=24)
plt.xlabel("xlabel", fontsize=16)
plt.ylabel("ylabel", fontsize=16)
plt.axis([0, 12, 0, 70]) # x: 0-8  y: 0-70
plt.tick_params(axis="both", labelsize=12, color="red")
plt.show()
```

```python
# remove x/y label
import matplotlib.pyplot as plt

data = [1,4,9,16,25,36,49,64]
plt.plot(data, linewidth=3)
plt.title("my title", fontsize=24)
plt.axis([0, 12, 0, 70]) # x: 0-8  y: 0-70
plt.tick_params(axis="both", labelsize=12, color="red")
plt.axes().get_xaxis().set_visible(False) #remove x label
plt.axes().get_yaxis().set_visible(False) #remove y label
plt.show()
```

```python
# assign the position for each data
import matplotlib.pyplot as plt

data = [1,4,9,16,25,36,49,64]
seq = [3,4,5,6,7,8,9,12]   #assign the position for each data
plt.plot(seq, data, linewidth=3)
plt.title("my title", fontsize=24)
plt.xlabel("xlabel", fontsize=16)
plt.ylabel("ylabel", fontsize=16)
plt.axis([0, 12, 0, 70]) # x: 0-8  y: 0-70
plt.tick_params(axis="both", labelsize=12, color="red")
plt.show()
```

```python
# multiple data
import matplotlib.pyplot as plt

data1 = [1,4,9,16,25,36,49,64] # multiple data
data2 = [1,2,3,4,5,6,7,8]      # multiple data
seq = [3,4,5,6,7,8,9,12]
plt.plot(seq, data1, seq, data2, linewidth=3)
plt.title("my title", fontsize=24)
plt.xlabel("xlabel", fontsize=16)
plt.ylabel("ylabel", fontsize=16)
plt.tick_params(axis="both", labelsize=12, color="red")
plt.show()
```

```python
# customized line's style and color
#b=blue c=cyan g=green k=black m=magenta r=red w=white y=yellow
#-=solid(default) --=dashed -.=dashdot :=dotted .=dot , o v ^ s p * + -
import matplotlib.pyplot as plt

data1 = [1,1,1,1,1]
data2 = [2,2,2,2,2]
data3 = [3,3,3,3,3]
data4 = [4,4,4,4,4]
data5 = [5,5,5,5,5]
data6 = [6,6,6,6,6]
data7 = [7,7,7,7,7]
seq = [1,2,3,4,5]   #assign the position for each data
plt.plot(seq, data1, "g--",
         seq, data2, "r-.",
         seq, data3, "y:",
         seq, data4, "k.",
         seq, data5, "m,",
         seq, data6, "co",
         seq, data7, "kv")
plt.title("my title", fontsize=24)
plt.xlabel("xlabel", fontsize=16)
plt.ylabel("ylabel", fontsize=16)
plt.tick_params(axis="both", labelsize=12, color="red")
plt.show()
```

### sin, cos

```python
import matplotlib.pyplot as plt
import numpy as np

xpt = np.linspace(0, 10, 100)
ypt1 = np.sin(xpt)
ypt2 = np.cos(xpt)
plt.scatter(xpt, ypt1, color=(0, 1, 0))
plt.scatter(xpt, ypt2)
plt.show()
```

```python
# lwdiths, point size
import matplotlib.pyplot as plt
import numpy as np

xpt = np.linspace(0, 5, 100)
lwdiths = xpt**3
plt.scatter(xpt, xpt, s=lwdiths, color=(0, 1, 0))
plt.show()
```

```python
# color=c, point color
import matplotlib.pyplot as plt
import numpy as np

xpt = np.linspace(0, 5, 100)
plt.scatter(xpt, xpt, c=xpt, cmap="rainbow")
plt.show()
```

### multi-graph

```python
import matplotlib.pyplot as plt

data1 = [x for x in range(1,9)]
data2 = [x*x for x in range(1,9)]
seq = data1

plt.figure(1)
plt.plot(seq, data1, "-*")

plt.figure(2)
plt.plot(seq, data2, "-o")
plt.title("title", fontsize=24)
plt.xlabel("x label", fontsize=14)
plt.ylabel("y label", fontsize=14)

plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
from datetime import datetime

with open("res/TaipeiWeatherJan.csv") as f:
    reader = csv.reader(f)
    next(reader) # skip tag
    date, high, mean, low = [], [], [], []
    for row in reader:
        date.append(datetime.strptime(row[0], "%Y/%m/%d"))
        high.append(int(row[1]))
        mean.append(int(row[2]))
        low.append(int(row[3]))

fig = plt.figure(dpi=80, figsize=(12,8)) #get figure
plt.plot(date, high)
fig.autofmt_xdate() #set x label rotation
plt.title("title", fontsize=24)
plt.xlabel("date", fontsize=14)
plt.ylabel("Temperature", fontsize=14)
plt.tick_params(axis="both", labelsize=10)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
from datetime import datetime

with open("res/TaipeiWeatherJan.csv") as f:
    reader = csv.reader(f)
    next(reader)
    date, high, mean, low = [], [], [], []
    for row in reader:
        date.append(datetime.strptime(row[0], "%Y/%m/%d"))
        high.append(int(row[1]))
        mean.append(int(row[2]))
        low.append(int(row[3]))

fig = plt.figure(dpi=80, figsize=(12,8))
plt.plot(date, high) # draw more line
plt.plot(date, mean) # draw more line
plt.plot(date, low)  # draw more line
plt.fill_between(date, high, low, color="y", alpha=0.1) # fill in the color
plt.fill_between(date, mean, low, color="r", alpha=0.1) # fill in the color
fig.autofmt_xdate()
plt.title("title", fontsize=24)
plt.xlabel("date", fontsize=14)
plt.ylabel("Temperature", fontsize=14)
plt.tick_params(axis="both", labelsize=10)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
from datetime import datetime

with open("res/TaipeiWeatherJan.csv") as f:
    reader = csv.reader(f)
    next(reader)
    date, high, mean, low = [], [], [], []
    for row in reader:
        date.append(datetime.strptime(row[0], "%Y/%m/%d"))
        high.append(int(row[1]))
        mean.append(int(row[2]))
        low.append(int(row[3]))

fig = plt.figure(dpi=80, figsize=(12,8))
plt.plot(date, high)
plt.plot(date, mean)
plt.plot(date, low)
plt.fill_between(date, high, low, color="y", alpha=0.1)
plt.fill_between(date, mean, low, color="r", alpha=0.1)
plt.axis([date[0], date[30], 0, 40])
plt.title("title", fontsize=24)
plt.xlabel("date", fontsize=14)
plt.ylabel("Temperature", fontsize=14)
plt.tick_params(axis="both", labelsize=10)
fig.autofmt_xdate()
plt.show()
```
